```python
import random
import numpy as np
np.float = float

import serial
from serial.tools import list_ports
from psychopy import visual, core, event, clock
import tkinter as tk
import time


'''
Open serial port and create function "vibrate" to send labels in EEG about picture 
'''

ser = serial.Serial(list_ports.comports()[0].device, 9600, timeout= 0.1)
def vibrate(message=1, ser=ser):
    ser.write(bytes(str(message), 'utf-8'))
    
# ser = 0 # for testing
# def vibrate(message=1, ser=ser):
#     pass
```

```python
root = tk.Tk()
w = root.winfo_screenwidth()
h = root.winfo_screenheight()
# stims = ["4.jpg", "left.jpg"]
stims_number = 50
total_stims_number = stims_number * 2
trial_time = 1.5
isi_time = 1 # interstimulus time
map_types_num_images = {0: [f"0 ({i}).jpeg" for i in range(1, 44)], 1: [f"{i}.jpeg" for i in range(1, 44)]}

txt_1 = u'Focus to your feelings at the picture'
txt_2 = u'Experiment is over\nFor exit click to the "space"'

win = visual.Window([w, h], color=(0, 0, 0), colorSpace='rgb255', fullscr=True, screen=1)
ISI = clock.StaticPeriod(win=win, screenHz=59, name='ISI')

####OPENING
txt = visual.TextStim(win, text= u'Quite Smart', font='Consolas', pos=[0, 0], color = (0,0,0))
txt.draw()
win.flip()
event.waitKeys(keyList=['space'])

####OBSERVATION BEFORE
txt = visual.TextStim(win, text=txt_1, font='Consolas', pos=[0, 0], color = (0,0,0))
txt.draw()
win.flip()
event.waitKeys(keyList=['space'])

coef = 2.0
fig_size = 1080

movs = {
    key : 
    [visual.ImageStim(
        win, 
        image=i, 
        mask=None, 
        units='', 
        pos=(0.0, 0.0), 
        size=(coef*fig_size/w, 
              coef*fig_size/h),
        ori=0.0,
        color=(1.0, 1.0, 1.0),
        colorSpace='rgb',
        contrast=1.0,
        opacity=1.0,
        depth=0,
        interpolate=False,
        flipHoriz=False,
        flipVert=False,
        texRes=128,
        name=None,
        autoLog=None,
        maskParams=None
    ) for i in stim] for key, stim in map_types_num_images.items()}
stim_types = [0]*stims_number + [1]*stims_number
# stim_types = np.arange(17) 
np.random.shuffle(stim_types)

for a, stim_type in zip(range(total_stims_number), stim_types):
    s0 = time.time()
    
    mov = random.choice(movs[stim_type])
    mov.draw()
    win.flip()
    
    if stim_type == 0: # stim_type - rest or movement
        vibrate(6)
    else:
        vibrate(7)
    
    core.wait(trial_time)
    
    s1 = time.time()
    print(s1-s0)     
    win.flip()
    ISI.start(isi_time)
    ISI.complete()

txt = visual.TextStim(win, text=txt_2, font='Consolas', pos=[0, 0], color=(0,0,0))
txt.draw()       
win.flip()
event.waitKeys(keyList=['space'])  
     
win.close()
core.quit()
```
