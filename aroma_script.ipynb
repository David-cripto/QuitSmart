```python
import requests
import socket
import serial
from serial.tools import list_ports
import tkinter as tk
import numpy as np
from psychopy import visual, core, event, clock
import time 
```

```python
URL = 'http://192.168.43.254:1003' # faster
```

```python
ser = serial.Serial(list_ports.comports()[0].device, 9600, timeout= 0.1)
def vibrate(message=1, ser=ser):
    ser.write(bytes(str(message), 'utf-8'))
```

```python
root = tk.Tk()
w = root.winfo_screenwidth()
h = root.winfo_screenheight()
win = visual.Window([w, h], color=(0, 0, 0), colorSpace='rgb255', fullscr=True, screen=1)
ISI = clock.StaticPeriod(win=win, screenHz=59, name='ISI')

trial_time = 2
question_time = 5
isi_time = 1 # interstimulus time
# number_of_blocks = 6
stims_number = 57
total_stims_number = 2 * stims_number
stim_types = [0]*stims_number + [1]*stims_number
np.random.shuffle(stim_types)

for a, stim_type in zip(range(total_stims_number), stim_types):
    s0 = time.time()
    
    txt = visual.TextStim(win, text='Breath now', font='Consolas', pos=[0, 0], color = (0,0,0))
    txt.draw()
    win.flip()
    json = {
        "channels": [1 if stim_types == 0 else 2],
        "intensities": [50],
        "durations": [int(trial_time) * 1000],
        "booster": True
    }
    r = requests.post(URL+'/as2/diffuse', json=json, headers={"Content-Type": "application/json"})
    
    if stim_type == 0: # stim_type - rest or movement
        vibrate(6) # апельсин
    else:
        vibrate(7) # лаванда
    
    core.wait(trial_time)
    
    txt = visual.TextStim(win, text='What did you smell?', font='Consolas', pos=[0, 0], color = (0,0,0))
    txt.draw()
    win.flip()
    
    core.wait(trial_time)
    
    s1 = time.time()
    print(s1-s0)
    win.flip()
    ISI.start(isi_time)
    ISI.complete()
    

txt = visual.TextStim(win, text=u'Experiment is over\nFor exit click to the "space"', font='Consolas', pos=[0, 0], color=(0,0,0))
txt.draw()       
win.flip()
event.waitKeys(keyList=['space'])  
     
win.close()
core.quit()
```
